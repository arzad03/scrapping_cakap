```python
from google_play_scraper import app, Sort
import pandas as pd
import numpy as np
from google_play_scraper import app, reviews_all
```

```python
result = reviews_all(
    'com.google.android.apps.bard',
    sleep_milliseconds=0,
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    filter_score_with=None
)
```

```python
df_awal = pd.DataFrame(np.array(result),columns=['review'])
df_awal = df_awal.join(pd.DataFrame(df_awal.pop('review').tolist()))
df_awal.head()
```

```python
scrappeddata1 = df_awal[['content','score','at']]
sorteddata = scrappeddata1.sort_values(by='at', ascending=True) #mengurutkan dari yang tebaru
sorteddata.head()
```

```python
sorteddata.to_csv('review_bard.csv') #write dataframe into csv file
```

```python
import requests
from bs4 import BeautifulSoup
from csv import writer
import csv
```

```python
from os import setgroups
url = "https://www.imdb.com/chart/moviemeter/"
response = requests.get(url) # kirimkan header
soup = BeautifulSoup(res.content, 'html.parser') # parsing html

movie_containers = soup.find_all('td', class_ = 'titleColumn')
```

```python
with open('/content/sample_data/list_movies.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Title', 'Year', 'Rating'])

    for container in movie_containers:
        title = container.a.text
        year = container.span.text

        rating_container = container.find_next_sibling('td', class_='ratingColumn imdbRating')
        rating = rating_container.text if rating_container else 'N/A'

        writer.writerow([title, year, rating])
```
